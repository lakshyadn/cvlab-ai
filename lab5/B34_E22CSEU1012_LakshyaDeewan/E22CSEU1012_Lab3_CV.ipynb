```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# Task-1: Image Manipulation
image = cv2.imread('/content/Picture1.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display Image
plt.imshow(image_rgb)
plt.title('Original Image')
plt.show()

# Extract Image Size
height, width, channels = image.shape
print(f'Image Size: {width}x{height}, Channels: {channels}')

# Calculate Total Pixels
total_pixels = height * width
print(f'Total Pixels: {total_pixels}')

# Convert to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow( gray_image)

# Convert to Binary using Threshold
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
cv2_imshow( binary_image)

# Count Black Pixels
black_pixel_count = np.sum(binary_image == 0)
print(f'Black Pixels Count: {black_pixel_count}')

# Task-2: Edge Detection
sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobelx, sobely)
cv2_imshow( sobel_combined)

prewittx = cv2.filter2D(gray_image, -1, np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]]))
prewitty = cv2.filter2D(gray_image, -1, np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]]))
prewittx = np.float32(prewittx)
prewitty = np.float32(prewitty)

prewitt_combined = cv2.magnitude(prewittx, prewitty)
cv2_imshow( prewitt_combined)

robertsx = np.array([[1, 0], [0, -1]])
robertsy = np.array([[0, 1], [-1, 0]])
robertsx_edge = cv2.filter2D(gray_image, -1, robertsx)
robertsy_edge = cv2.filter2D(gray_image, -1, robertsy)
robertsx_edge = np.float32(robertsx_edge)
robertsy_edge = np.float32(robertsy_edge)
roberts_combined = cv2.magnitude(robertsx_edge, robertsy_edge)
cv2_imshow( roberts_combined)

canny_edges = cv2.Canny(gray_image, 100, 200)
cv2_imshow(canny_edges)

# Task-2: Edge Detection
sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobelx, sobely)
cv2_imshow(sobel_combined)

prewittx = cv2.filter2D(gray_image, cv2.CV_64F, np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]]))
prewitty = cv2.filter2D(gray_image, cv2.CV_64F, np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]]))
prewitt_combined = cv2.magnitude(prewittx, prewitty)
cv2_imshow(prewitt_combined)

robertsx = np.array([[1, 0], [0, -1]])
robertsy = np.array([[0, 1], [-1, 0]])
robertsx_edge = cv2.filter2D(gray_image, cv2.CV_64F, robertsx)
robertsy_edge = cv2.filter2D(gray_image, cv2.CV_64F, robertsy)
roberts_combined = cv2.magnitude(robertsx_edge, robertsy_edge)
cv2_imshow(roberts_combined)

canny_edges = cv2.Canny(gray_image, 100, 200)
cv2_imshow(canny_edges)
```
